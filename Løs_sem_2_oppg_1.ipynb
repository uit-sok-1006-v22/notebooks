## 7.2 Tolkning av tilbud og etterspørsel

```python
# definer symboler
q, p =sp.symbols('q p')

# løs slik at vi får logQ= funksjon av logP
# La q=logQ og p=logP
# Tilbud p=-2+1.7q (i log)
# Etterspørsel p=8.5-0.82q (i log)

def tilbud(q):
    return -2+1.7*q
sp.solve(p-tilbud(q),q)[0]
```

Da har vi $\frac {dlogQ}{dlogP} =0.588$ som er tilbudselastisiteten.
Husk $d(logQ)=\frac{1}{Q}dQ$ og $d(logP)=\frac{1}{P}dP$.    
Slik at $\frac{dlogQ}{dlogP}=\frac{\frac{1}{Q}dQ}{\frac{1}{P}dP} = \frac {dQ}{Q}\frac{P}{dP}= \frac{dQ}{dP}\frac{P}{Q}$ som er elastisitetsformelen!

```python
def etterspørsel(q):
    return 8.5-0.82*q
sp.solve(p-etterspørsel(q),q)[0]
```

Etterspørselselastisiteten er følgelig -1.2195.
