```python
# importer pakkene som vi trenger
import numpy as np
from matplotlib import pyplot as plt
import sympy as sp
```

```python
# definer symboler
N,K,a,b,Z=sp.symbols('N K a b Z', positive=True)
```

```python
def prod(N,K,a,b,Z):
    return Z*N**a*K**b
prod(N,K,a,b,Z)
```

#### deloppgave a)

Produksjonselastisiteten til arbeidskraft er gitt ved $f'_{N}\frac{N}{x}$. Vi skal bekrefte at dette er a.

```python
def elast_N(N,K,a,b,Z):
    return sp.diff(prod(N,K,a,b,Z),N)*(N/prod(N,K,a,b,Z))
elast_N(N,K,a,b,Z)
```

#### deloppgave b)

```python
def elast_K(N,K,a,b,Z):
    return sp.diff(prod(N,K,a,b,Z),K)*(K/prod(N,K,a,b,Z))
elast_K(N,K,a,b,Z)
```

#### deloppgave c)    
Skalaegenskaper. La oss tenke oss at vi fordobler N og K, og undersøk resulterende produksjon i forhold til 2 ganger opprinnelig mengde.

```python
def prod_2(N,K,a,b,Z):
    return Z*(2*N)**a*(2*K)**b
prod_2(N,K,a,b,Z)
```

```python
def overskudd(N,K,a,b,Z):
    return prod_2(N,K,a,b,Z)-2*prod(N,K,a,b,Z)
overskudd(N,K,a,b,Z)
```

```python
sp.simplify(overskudd(N,K,a,b,Z))
```

Vi kan se resultatet direkte herfra ettersom fortegnet til dette uttrykket er kun avhengig av leddet i parentes $(2^{a+b}-2)$.    
Vi ser at dette er 0 dersom $a+b=1$, større enn 0 dersom $a+b>1$, og mindre enn null om $1>a+b>0$.    
Konklusjon: $a+b=1$ gir konstant skalautbytte, $a+b>1$ stigende SUB, og $1>a+b>0$ avtakende SUB.

Merk at ingenting her avhenger av at vi fordoble innsatsfaktorene. Vi kan t-doble og få samme resultat. La oss vise dette for t>1:

```python
t=sp.symbols('t')
def prod_t(N,K,a,b,Z,t):
    return Z*(t*N)**a*(t*K)**b
prod_t(N,K,a,b,Z,t)
```

```python
def overskudd_t(N,K,a,b,Z,t):
    return prod_t(N,K,a,b,Z,t)-t*prod(N,K,a,b,Z)
overskudd_t(N,K,a,b,Z,t)
```

```python
sp.simplify(overskudd_t(N,K,a,b,Z,t))
```

Igjen er fortegnet til dette uttrykket kun avhengig av leddet i parentes $(t^{a+b}-t)$, med samme konklusjon som før ettersom t>1.
