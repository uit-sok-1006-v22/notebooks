```python
# importer pakkene som vi trenger
import numpy as np
from matplotlib import pyplot as plt
import sympy as sp
```

```python
x,w,r, N,K=sp.symbols('x w r,N K', real=True, positive=True)
```

```python
def C(w,r,x):
    return 2*(w*r*x)**0.5
C(w,r,x)
```

### deloppgave a)

```python
# Bruk Shephards Lemma
def Arb(w,r,x):
    return sp.diff(C(w,r,x),w)
def Kap(w,r,x):
    return sp.diff(C(w,r,x),r)
display(Arb(w,r,x))
display(Kap(w,r,x))
```

Da har vi    
$N(w,r,x)=\frac{{(rwx)}^{0.5}}{w}$    
$K(w,r,x)=\frac{{(rwx)}^{0.5}}{r}$.

### deloppgave b)

```python
#LÃ¸s for w
sp.solve(Arb(w,r,x)-N,w)
```

Da har vi $w=\frac{rx}{N^2}$

```python
# sett inn i kapitallikningen
Kap((r*x)/N**2,r,x)
```

### deloppgave c)
Som sier $K=\frac{2x}{N}$, dvs $x=\frac{KN}{2}$.    
Dette er produktfunksjonen!    
Cobb-Douglas av formen $x=ZN^aK^b$ med $Z=\frac{1}{2}, a=1, b=1$.

```python
# Sjekk at dette stemmer med kostnadsfunksjonen
Kost=w*Arb(w,r,x)+r*Kap(w,r,x)
Kost
```
